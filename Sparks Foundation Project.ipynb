```python
#Importing important libraries and packages.
# Pandas to import and analyze data
import pandas as pd
#Seaborn and Matplotlib for visualising data.
import seaborn as sns
import matplotlib.pyplot as plt
#Sklearn model is use to import required Algorithms/models.
#To split data into train and test sets we import train_test_split()
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
#Here we are going to import Linear Regression model.
from sklearn.linear_model import LinearRegression
```

```python
# Calling out given data set.
dataset = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
print(dataset)
```

```python
#By Use of seaborn and matplotlib we can visualize the data and find out correlation between two variables.
#Scatterplot
sns.scatterplot(x=dataset['Hours'], y=dataset['Scores'])
plt.title('Hours and Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
#Regressionplot
sns.regplot(x=dataset['Hours'], y=dataset['Scores'])
plt.title('Hours and Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
#Regressionplot
sns.regplot(x=dataset['Hours'], y=dataset['Scores'])
plt.title('Hours and Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
#Checking Correlation within given two variables.
print(dataset.corr())
```

```python
# Training Data
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
#Spliting the data into train and test sets.
#X-train and Y-train are used to train the model.
#X-test and Y-test are used to test the moedl.
#Test size distributes the data by 20% to each variable.
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
```

```python
#Training the algorithm.
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Predicting the output
pred_y = regressor.predict(X_test)
prediction = pd.DataFrame({'Hours': [i[0] for i in X_test], 'Predicted Marks': [k for k in pred_y]})
print(prediction)
```

```python
#Calculating mean square error
MAE = mean_absolute_error(X_test,pred_y)
print('Mean Absolute Error', MAE)
```
